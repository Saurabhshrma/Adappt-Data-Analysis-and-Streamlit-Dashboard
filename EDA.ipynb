```python
# Converting Datetime column to Datetime object

def convert_datetime(datetime_string):
  # This function converts a string to a datetime object.
  try:
    return dt.datetime.strptime(datetime_string, '%Y-%m-%d %H:%M:%S')
  except ValueError:
    return None

data['Datetime'] = data['Datetime'].apply(convert_datetime)
```

```python
#Load the datasets
details_df = pd.read_csv("Data/details.csv")
timeseries_df = pd.read_csv("Data/data.csv")
```

```python
# Convert the "Datetime" column to datetime type
timeseries_df["Datetime"] = pd.to_datetime(timeseries_df["Datetime"])
```

```python
# Merge details_df with timeseries_df to get timezone offset
combined_df = pd.merge(timeseries_df, details_df[["sensorId", "timezoneOffset"]], on="sensorId")
```

```python
# Convert the "Datetime" column to datetime type
combined_df["Datetime"] = pd.to_datetime(combined_df["Datetime"])

# Adjust the timezone offset
combined_df["Datetime"] = combined_df["Datetime"] + pd.to_timedelta(combined_df["timezoneOffset"], unit="ms")
```

```python
## Getting the Data with 10 min interval 

df = combined_df.copy()

# Convert the Datetime column to a datetime index
df["Datetime"] = pd.to_datetime(df["Datetime"])

# Set the frequency of the Datetime column to 10 minutes
df.set_index("Datetime", inplace=True)

# Group the DataFrame by sensor ID and 10 minute interval
grouped_df = df.groupby(["sensorId", pd.Grouper(level="Datetime", freq="10T")])["peopleCount"].sum()

```

```python
grouped_df = grouped_df.reset_index()
grouped_df
```

```python
combined_df = pd.merge(grouped_df, details_df, on="sensorId")
combined_df
```

```python
# Treating the outliers

# Calculate the IQR
Q1 = combined_df["peopleCount"].quantile(0.25)
Q3 = combined_df["peopleCount"].quantile(0.75)
IQR = Q3 - Q1

# Define the outlier threshold
outlier_threshold = 1.5 * IQR

# Filter the data to remove outliers
filtered_df = combined_df[
    (combined_df["peopleCount"] >= Q1 - outlier_threshold) &
    (combined_df["peopleCount"] <= Q3 + outlier_threshold)
]

filtered_df.shape
```
